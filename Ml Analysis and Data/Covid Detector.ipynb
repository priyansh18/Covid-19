```python
data=[['1', int(final.iloc[:,-11])], ['2', int(final.iloc[:,-10])], ['3', int(final.iloc[:,-9])], ['4', int(final.iloc[:,-8])], ['5', int(final.iloc[:,-7])], ['6', int(final.iloc[:,-6])], ['7', int(final.iloc[:,-5])], ['8', int(final.iloc[:,-4])], ['9', int(final.iloc[:,-3])], ['10', int(final.iloc[:,-2])] ]
```

```python
df = pd.DataFrame(data, columns = ['Day', 'Cases']) 
df
```

```python
x =np.asarray(df['Day']).reshape(-1,1)
y =np.asarray(df['Cases']).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
predict_test = model.predict(x_test)
```
